```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Books_Data_Clean.csv',index_col=False)
df = df.drop(columns=['index'])  

df.head()
```

```python
df.describe()
```

```python
df=df[df['Publishing Year']>1900]
```

```python
df.isna().sum()
```

```python
df.dropna(subset='Book Name',inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
plt.hist(df['Publishing Year'])
plt.xlabel('Publishing Year')
plt.ylabel('Frequency')
plt.title('distributing of Publishing Years')
plt.show()
```

```python
df['genre'].value_counts().plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('No of Books')
plt.title('No of books in each Genre')
plt.show()
```

```python
df.groupby("Author")['Book_average_rating'].mean().sort_values(ascending=False)
```

```python
sns.boxplot(x='genre',y='Book_ratings_count',data=df)
plt.xlabel('Genre')
plt.ylabel(' Books rating counts')
plt.title('Box plot of book rating counts for each genre ')
plt.show()
```

```python
Q1 = df['Book_ratings_count'].quantile(0.25)
Q3 = df['Book_ratings_count'].quantile(0.75)
IQR = Q3 - Q1

# Step 2: Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 3: Find outliers
outliers = df[(df['Book_ratings_count'] < lower_bound) | (df['Book_ratings_count'] > upper_bound)]

# Step 4: How many books have outliers
print("Number of outlier books:", outliers.shape[0])

```
